## Groupby
Split -> Apply -> Combine   
Split: Index가 같은 놈들끼리 묶어주기    
Apply: sum, mean, std 같은 함수를 적용시켜 줌   
Combine: Apply된 애들을 합쳐서 하나의 결과로 보여줌   

```python
df = pd.read_excel("input_file/titanic.xlsx")
df.head(2)
```

```python
# Embarked: 묶음의 기준
# Fare: sum의 적용 받는 column

df.groupby("Embarked")["Fare"].sum()
```

```python
# hierarchical index - 두 개의 column으로 groupby 

h_index = df.groupby(["Embarked","Sex"])["Fare"].sum()    # excel의 pivot tabel과 동일
h_index
```

```python
h_index.index
```

```python
# hierarchical index 형태를 matrix 형태로 다시 변환

h_index.unstack()
```

```python
h_index.unstack().stack()
```

```python
h_index.reset_index()
```

```python
# Embarked와 Sex의 level이 바껴서 나타남

h_index.swaplevel()
```

```python
h_index.sort_index(level=1)
```

```python
# grouped: groupby에 의해 split된 상태를 추출 가능

grouped = df.groupby("Embarked")

for name, group in grouped: 
    print(name)
    print(group)
```

```python
grouped.get_group("S")
```

```python
# aggregation

grouped.agg(sum)
```

```python
grouped.agg(np.mean)
```

```python
grouped['Fare'].agg([np.sum, np.mean, np.std])
```

```python
grouped.describe()
```

```python
# transformation

grouped = df.groupby("Fare")

score = lambda x: (x+1)
grouped.transform(score)
```

```python
score = lambda x: (x.max())
grouped.transform(score)
```

```python
score = lambda x: (x - x.mean()) / x.std()  # normalization
grouped.transform(score)
```

```python
# filter

df.groupby("Embarked").filter(lambda x: x["Fare"].sum() >= 15000)
```

```python
df.groupby("Pclass").filter(lambda x: x["Fare"].sum() >= 15000)
```
